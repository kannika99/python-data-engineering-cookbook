```python
def first_unique_char(ip_string):
    
    char_count = {}
    
    for i in ip_string:
        char_count[i] = char_count.get(i,0)+1 #dictionary.get(key, default_value)
        #print(char_count)
        
        
    print("Final Dict: ", char_count)
    
    for char in ip_string:
        if char_count[char] ==1:
            return char
        
    return -1
```

```python
first_unique_char("leetcode")
```

```python
first_unique_char("kannika")
```

```python
first_unique_char("kanniika")
```

```python
#Approach 1: Brute Force - O(n^2) - loop inside loop

def sum_nums(ip_list, target):
    
    for i in range(len(ip_list)):
        for j in range(i+1,len(ip_list)):
            if ip_list[i] + ip_list[j] ==target:
                return [i,j]
            
    return []    
    
```

```python
print(sum_nums([2, 7, 11, 15], 9))
```

```python
print(sum_nums([3, 2, 4], 6))       # Output: [1, 2]
print(sum_nums([3, 3], 6))          # Output: [0, 1]
```

```python
#Approach 2: Hashmap - O(n) - single for loop

def sum_nums_dict(ip_list, target):
    
    map_dict = {} # Dictionary (hashmap) to store numbers and their indices
    
    for i, num in enumerate(ip_list):
        complement = target - num # Find the number needed to reach the target
        
        print(num, complement)
        
        if complement in map_dict:  # Check if the complement exists in the dictionary
            return [map_dict[complement],i]
        
        map_dict[num] = i
        
    
    return []
```

```python
sum_nums_dict([2, 7, 11, 15], 9)
```

```python
print(sum_nums([3, 2, 4], 6))       # Output: [1, 2]
print(sum_nums([3, 3], 6))    
```
