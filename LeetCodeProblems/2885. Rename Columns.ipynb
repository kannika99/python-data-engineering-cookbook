Q no. 2885. Rename Columns

Write a solution to rename the columns as follows:

id to student_id

first to first_name

last to last_name

age to age_in_years

The result format is in the following example.

```python
#Solution

import pandas as pd

def renameColumns(studentdf):
    return studentdf.rename(columns={'id':'student_id', 'first':'first_name', 'last':'last_name','age':'age_in_years'})
```

```python
students = pd.DataFrame({'id':[1,2,3,4,5], 
                        'first':['Mason','Ava','Taylor','Georgia','Thomas'],
                        'last':['A','B','Sharma','Moore', 'k'],
                        'age':[6,19,25,8,32]})
print(students)
```

```python
new_studentsdf = renameColumns(students)
```
